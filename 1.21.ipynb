#### Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

### Load the data

```python
data = pd.read_excel("C:\\Users\\kwame\\Downloads\\1.21.xlsx",  names=["broken", "transfer"])
```

#### a. Perform linear regression and obtain bo and b1

```python
X = data["transfer"]
y = data["broken"]
X = sm.add_constant(X)  # Add a constant for the intercept
model = sm.OLS(y, X).fit()
b0, b1 = model.params["const"], model.params["transfer"]
print("b0 =", b0)
print("b1 =", b1)
```

#### b. Plot the regression line and data

```python
plt.scatter(data["transfer"], data["broken"])
plt.plot(data["transfer"], model.predict(X), color="blue")
plt.xlabel("Number of Transfers")
plt.ylabel("Number of Broken Ampules")
plt.show()
```

#### c. Estimate Y when X = 1 transfer

```python
X_value = 1
estimated_Y = b0 + b1 * X_value
print("Estimated Y for X = 1 transfer is:", estimated_Y)
```

#### d. Estimate the increase in Y when there are 2 transfers compared to 1 transfer

```python
increase_in_Y = b1
print("Estimated increase in Y for a 1-unit increase in transfers:", increase_in_Y)
```

#### e.Verify the line goes through bar(X) and bar(Y)

```python
Y_bar = data["broken"].mean()
X_bar = data["transfer"].mean()
predicted_Y_bar = b0 + b1 * X_bar
print("Mean of Y:", Y_bar)
print("Predicted mean of Y based on the regression line:", predicted_Y_bar)
```
